#Use Cases
1) Use Case: Display tweets searched by a keyword "Job".
Description: User can view tweets containing the job keyword

Actor: User

Precondition: When a user wants to search tweets using a specific keyword. 

Steps: 'Select * from twitter_bot where Tweet like "%Job%" ', con = conn

Actor action: User views tweets containing keyword Job.

System Responses: The list of tweets with the keyword Job is displayed.

Post Condition:User successfully views the tweets with the keyword Job.

Alternate Path: The are no tweets with keyword Job available to display.

Error: Tweets searched by keyword "Job" do not exist.
